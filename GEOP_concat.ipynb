```python
import xarray as xr
import numpy as np
import datetime
import matplotlib.pyplot as plt
import pandas as pd
# import cartopy.crs as ccrs
# import cartopy.feature as cfeature
# from cartopy.util import add_cyclic_point
import math
from collections import Counter
```

```python
ene = 'geopotential_01.nc'
feb = 'geopotential_02.nc'
mar = 'geopotential_03.nc'
abr = 'geopotential_04.nc'
may = 'geopotential_05.nc'
jun = 'geopotential_06.nc'
jul = 'geopotential_07.nc'
ago = 'geopotential_08.nc'
sep = 'geopotential_09.nc'
octb = 'geopotential_10.nc'
nov = 'geopotential_11.nc'
dic = 'geopotential_12.nc'
```

```python
ds_ene = xr.open_dataset(ene)
ds_feb = xr.open_dataset(feb)
ds_mar = xr.open_dataset(mar)
ds_abr = xr.open_dataset(abr)
ds_may = xr.open_dataset(may)
ds_jun = xr.open_dataset(jun)
ds_jul = xr.open_dataset(jul)
ds_ago = xr.open_dataset(ago)
ds_sep = xr.open_dataset(sep)
ds_oct = xr.open_dataset(octb)
ds_nov = xr.open_dataset(nov)
ds_dic = xr.open_dataset(dic)
```

```python
dsene = ds_ene['z']
dsfeb = ds_feb['z']
dsmar = ds_mar['z']
dsabr = ds_abr['z']
dsmay = ds_may['z']
dsjun = ds_jun['z']
dsjul = ds_jul['z']
dsago = ds_ago['z']
dssep = ds_sep['z']
dsoctb = ds_oct['z']
dsnov = ds_nov['z']
dsdic = ds_dic['z']
```

```python
gph500_daily = xr.concat([dsene, dsfeb,dsmar,dsabr,dsmay,dsjun,dsjul,dsago,dssep,dsoctb,dsnov,dsdic], dim='time')
```

```python
# Guardar el dataset en un archivo NetCDF
gph500_daily.to_netcdf("gph500_daily2019-2024.nc")
```
